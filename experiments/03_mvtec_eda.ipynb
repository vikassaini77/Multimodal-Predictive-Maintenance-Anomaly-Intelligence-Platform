# MVTec EDA

Defect type distribution and sample grid visualization using `torchvision.utils.make_grid`.

```python
import torch
import torchvision
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import sys

# Add backend to path to import app modules
sys.path.append(str(Path.cwd().parent / 'backend'))
from app.data.mvtec_dataset import MVTecDataset
from app.data.mvtec_transforms import get_test_transforms

def show_grid(tensor_img):
    tensor_img = tensor_img.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    tensor_img = std * tensor_img + mean
    tensor_img = np.clip(tensor_img, 0, 1)
    plt.imshow(tensor_img)
    plt.axis('off')
```

```python
try:
    dataset = MVTecDataset(root_dir="../backend/data/mvtec", category="bottle", split="test", transform=get_test_transforms())
    
    images = [dataset[i][0] for i in range(min(4, len(dataset)))]
    if images:
        grid = torchvision.utils.make_grid(images)
        plt.figure(figsize=(10, 5))
        show_grid(grid)
        plt.title("MVTec 'bottle' Test Samples")
        plt.show()
    else:
        print("No images found. Please download dataset first.")
except FileNotFoundError:
    print("MVTec dataset not found. Run downloader.")
```
