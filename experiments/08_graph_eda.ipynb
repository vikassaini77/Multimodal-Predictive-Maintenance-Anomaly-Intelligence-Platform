# Factory Graph Topology EDA
Generates a synthetic factory topology, calculates centrality metrics, and visualizes the graph.

```python
import sys
sys.path.append('..')

import networkx as nx
import matplotlib.pyplot as plt
import torch
from backend.app.data.factory_graph import build_synthetic_factory_graph, convert_to_pyg_heterodata
from backend.app.utils.graph_visualizer import visualize_factory_graph
```

## Generate Graph

```python
# Generate graph with 3 assembly lines and 4 machines per line
factory_graph = build_synthetic_factory_graph(num_lines=3, machines_per_line=4)
print(f"Total Nodes: {factory_graph.number_of_nodes()}")
print(f"Total Edges: {factory_graph.number_of_edges()}")
```

## Graph Visualization

```python
# Let's highlight machines 2 and 7 as anomalous
anomalous_machines = ['machine_2', 'machine_7']
visualize_factory_graph(factory_graph, anomalous_nodes=anomalous_machines)
```

## NetworkX Topology Stats

```python
# Node degree distribution
degrees = [d for n, d in factory_graph.degree()]
plt.hist(degrees, bins=range(max(degrees)+2), align='left')
plt.title('Node Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.show()
```

```python
# Centrality Metrics
centrality = nx.degree_centrality(factory_graph)
# Show top 5 central nodes
sorted_centrality = sorted(centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 most central nodes:")
for node, score in sorted_centrality[:5]:
    print(f"{node}: {score:.4f}")
```

## PyG HeteroData Conversion

```python
# Provide mock 256-dim embeddings for 12 machines (3 lines * 4 machines)
machine_embeddings = torch.randn((12, 256))
hetero_data = convert_to_pyg_heterodata(factory_graph, machine_embeddings)
print(hetero_data)
print(f"Machine features shape: {hetero_data['machine'].x.shape}")
print(f"Feeds_into edges shape: {hetero_data[('machine', 'feeds_into', 'conveyor')].edge_index.shape}")
```
