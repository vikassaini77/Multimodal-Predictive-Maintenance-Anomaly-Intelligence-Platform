# ReAct Agent: Full Diagnosis-to-Alert Scenario

This notebook demonstrates the end-to-end trace of the ReAct agent diagnosing a fault, fetching visual confirmation, and attempting to dispatch an alert (which hits the safety guardrail).

```python
from backend.app.agent.react import ReActAgent
from backend.app.agent.registry import PermissionScope

# We instantiate with ACTION scope so it can dispatch alerts
agent = ReActAgent(agent_scope=PermissionScope.ACTION, max_iterations=7)

query = "The operator reported a strange noise from machine_001 around 10:10 AM today. Can you diagnose what's happening, check the camera snapshot, and dispatch a critical alert to the maintenance team if needed?"

result = agent.run(query)

print("\n=== AGENT TRACE ===")
for step in result["trace"]:
    print(step)

print("\n=== FINAL OUTPUT ===")
print(result["answer"])
```
