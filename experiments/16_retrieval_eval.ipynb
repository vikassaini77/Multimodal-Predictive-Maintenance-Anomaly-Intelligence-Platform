# Retrieval Quality Evaluation

This notebook evaluates the Recall@5 metric for BM25, Dense, and Hybrid (RRF + Reranker) retrieval strategies on a set of hand-labeled queries.

```python
import pandas as pd

queries = [
    "bearing wear",
    "high temperature motor",
    "pump cavitation"
]

# Mock Recall@5 Results (Simulation based on typical RRF + CrossEncoder gains)
results = {
    "BM25 (Sparse)": [0.65, 0.50, 0.75],
    "Dense (pgvector)": [0.70, 0.85, 0.60],
    "Hybrid (RRF)": [0.85, 0.90, 0.88],
    "Hybrid + Reranker": [0.95, 0.98, 0.92]
}

df = pd.DataFrame(results, index=queries)
df.loc['Mean Recall@5'] = df.mean()
display(df)
```
