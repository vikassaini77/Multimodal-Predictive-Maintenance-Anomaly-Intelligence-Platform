# RAG Corpus EDA

This notebook evaluates the text chunks and embeddings generated for the OEM maintenance manuals.

```python
from backend.app.rag.corpus import generate_synthetic_manuals
from backend.app.rag.chunker import TextChunker
from backend.app.rag.embeddings import EmbeddingGenerator
import numpy as np
import pandas as pd

docs = generate_synthetic_manuals()
chunker = TextChunker(chunk_size=512, overlap=50)
embedder = EmbeddingGenerator()

all_chunks = []
for d in docs:
    all_chunks.extend(chunker.chunk_document(d))
    
print(f"Total Documents: {len(docs)}")
print(f"Total Chunks Generated: {len(all_chunks)}")
```

```python
texts = [c['content'] for c in all_chunks]
embeddings = embedder.embed_batch(texts)

norms = np.linalg.norm(embeddings, axis=1)
df = pd.DataFrame({
    "Chunk Length (chars)": [len(t) for t in texts],
    "Embedding Norm": norms
})
display(df.describe())
```
