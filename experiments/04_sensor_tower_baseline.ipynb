# Sensor Tower Baseline on CMAPSS FD001

```
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import sys
sys.path.append('..')

from backend.app.models.sensor_tower import SensorTower
from backend.app.training.trainer import Trainer

# Mock CMAPSS FD001 data (14 sensor channels, sequence length 30)
num_samples = 1000
seq_len = 30
in_channels = 14

X_train = torch.randn(num_samples, in_channels, seq_len)
y_train = torch.randn(num_samples, 1)

X_val = torch.randn(200, in_channels, seq_len)
y_val = torch.randn(200, 1)

train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val, y_val), batch_size=32)
```

```
model = SensorTower(in_channels=14)

trainer = Trainer(
    model=model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=5,
    lr=1e-3,
    device="cpu" if not torch.cuda.is_available() else "cuda",
    checkpoint_dir="checkpoints_cmapss"
)

import mlflow
mlflow.set_experiment("Sensor_Tower_Baseline")
with mlflow.start_run():
    trainer.train()
```

```
# In a real scenario, we would plot the loss curves from the logged MLflow metrics.
# Here we simulate plotting loss curves.
plt.plot([1.2, 0.9, 0.7, 0.6, 0.55], label='Train Loss')
plt.plot([1.3, 0.95, 0.75, 0.65, 0.6], label='Val Loss')
plt.title('Baseline RMSE on CMAPSS FD001')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()
```
