# ReAct Agent Trace Evaluation

This notebook demonstrates the end-to-end trace of the ReAct agent as it diagnoses a fault using Gemini 2.5 Flash's function calling.

```python
from backend.app.agent.react import ReActAgent
from backend.app.agent.tools.sensor_tool import QuerySensorDBTool
from backend.app.agent.tools.anomaly_tool import RunAnomalyScoreTool
from backend.app.agent.tools.rag_tool import SearchManualRAGTool

tools = [
    QuerySensorDBTool(),
    RunAnomalyScoreTool(),
    SearchManualRAGTool()
]

agent = ReActAgent(tools=tools, max_iterations=5)

query = "The operator reported a strange noise from machine_001 around 10:10 AM today (2023-10-27). Can you diagnose what's happening and tell me what to do?"

result = agent.run(query)

print("\n=== AGENT TRACE ===")
for step in result["trace"]:
    print(step)

print("\n=== FINAL DIAGNOSIS ===")
print(result["answer"])
```
